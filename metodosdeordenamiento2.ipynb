metofo de ordenamiento de la burbuja (bubbke sort).

es un metodo de ot¿rdenamiento simple y funciona revisando cada elemento de la lista ordenar con el que sif¿guem canbiandilios de posiciom si estam em um ordem con el que sigue, cambiandolos de posicion si estan en un orden incorrecto (n>n+1), es necesario repetir este proceso varias veces hasta que no se necesiten ams cambios, lo que significa que la lista quedo ordenada

https://www.youtube.com/watch?v=lB-OJcTkgEs


```python
from random import sample # Importamos metodo random para generar las listas aleatorias

lista = list(range(100))# Creamos la lista de base con números del 1 al 100

# Creamos una lista aleatoria con sample
#(8 elementos aleatorios de la lista base)
vectorbs = sample(lista,8)# Selecciones los 8 elementos aletorios de la base de datos

# Vamos a estructurar la función para el método de ordenamiento

def bubblesort(vectorbs):
  """ Esta función ordena el vector que le pases como argumento con el Método Bubble Sort"""
  # Imprimimos la lista obtenidad al principio de forma desordenada
  print("El vector a ordenar es :",vectorbs)
  n = 0 # Establecemos un contador de largo del vector

  for _ in vectorbs:
      n += 1 # Contamos la cantidad de caracteres dentro del vector

  for i in range(n-1):
  # Le damos un rango n para que complete el proceso
      for j in range(0, n-i-1):
  # Revisa la matriz de 0 hasta n-i-1
          if vectorbs[j] > vectorbs[j+1]:
              vectorbs[j], vectorbs[j+1]= vectorbs[j+1], vectorbs[j]
  # Se intercambian si el elemento encontrado es mayor
  # Luego pasa al siguiente
  print ("El vector ordenado es :", vectorbs)

bubblesort(vectorbs)
```

metodo de ordenamiento de incersion (insertionsort)

actua recorriendo la lista a ordenas, tomando el elemento actual e incertandolo donde deberia, comprobandolo entre los que ya ha recorrido

[Método de Ordenamiento por Inserción](https://www.youtube.com/watch?v=sO-ix3EFrhA)

```python
from random import sample

lista = list(range(100))

vectorins = sample(lista,15)

def insertionsort(vectorins):
  """Esta función ordena el vector que le pases como argumento con el Método de Inserción sort"""

  # Imprimir la lista obtenida al principio (Desordenada)
  print("El vector a ordenar con Inserción es :", vectorins)

  largo = 0 # Establecemos el contador de largo

  for i in vectorins:
    largo += 1 # Obtenemos el largo del vector

# Recorremos la lista de 1 hasta el largo del vector
  for i in range(1, largo):

    elemento = vectorins[i]

    # Movemos los elementos del vectorins[0...i-1], que son mayores que el elemento
    # a una posición adelante de su posición actual

    j = i-1
    while j >= 0 and elemento < vectorins[j]:
      vectorins[j+1]=vectorins[j]
      j -= 1
    vectorins[j+1] = elemento
  print("El vector ordenado con inserción es: ", vectorins)

insertionsort(vectorins)
```

Metodo de ordenamiento shell

este es una mejora del metodo de ordenamiento por insercion, ya que el metodo de insercion es eficiente si la lista esta casi ordenada, para ello el metodo shell compara elementos separados por un espacio de varias posiciones y esto permite que el elemento haga "pasos mas grandes" hacia su posicion esperada, el mismo finaliza con un ordenamiento por insercion simple

[Método de Ordenamiento Shell Sort](https://www.youtube.com/watch?v=bJ-LWnpyx6s)

```python
from random import sample
lista = list(range(100))
vectorshell = sample(lista, 40)

def shellsort(vectorshell):
  """esta funcion ordena el vector que le pases como argumento con el metodo shell sort"""
  print("el vector a ordenar con shell es: ", vectorshell)

  largo = 0

  for i in vectorshell:
    largo += 1

  distancia = largo // 2

  #creamos un bucle segun las distancias
  while distancia > 0:
    #utilizamos el insertionsort
    for i in range(distancia, largo):
      valor = vectorshell[i]
      j = i
      while j >= distancia and vectorshell[j - distancia] > valor:
        vectorshell[j] = vectorshell[j - distancia]
        j -= distancia
      vectorshell[j] = valor
    distancia //= 2 # acortamos la distancia nuevamente y continuamos el ciclo
  print("el vector ordenado con shell es: ", vectorshell)

shellsort(vectorshell)
```

Metodo de ordenamiento por mezcla (mergosort)

este tiene un funcionamiento muy particular, primero debemos saber que si la longitud de la lista es 0 o 1 ya esta ordenado, en otro caso:
el algoritmo debera dividir la lista desordenada en dos sublistas de aproximadamente la mitad del tamaño, luego ordenara cada sublista recursivamente aplicando el ordenamiento por mezcla, y por ultimo mezcla las dos sublistas en una sola lista ordenada

[Método de Ordenamiento por Mezcla (Merge Sort)](https://www.youtube.com/watch?v=slMrT6hpDiQ)

```python
from random import sample

lista = list(range(100))
vectormerge = sample(lista, 30)

def mergesort(vectormerge):
  """Esta función ordena el vector que le pases como argumento con el método merge sort"""

  # Imprimimos la lista obtenida al principio (Desordenada)
  print("El vector a ordenar con merge es: ", vectormerge)

  def merge(vectormerge):

     def largo(vec):
      largovec = 0 # Establecemos un contador del largovec
      for _ in vec:
        largovec += 1 # Obtenemos el largo del vector
      return largovec

     if largo(vectormerge) > 1:
      medio = largo(vectormerge)//2 # Buscamos el medio del vector

      # lo dividimos en 2 partes
      izq = vectormerge[:medio]
      der = vectormerge[medio:]

      merge(izq)# Mismo procedimiento a la primer mitad
      merge(der)# Mismo procedimiento a la segunda mitad

      i=j=k=0

      # Copiamos los datos a los vectores temporales izq[] y der[]

      while i < largo(izq) and j < largo(der):
        if izq[i]<der[j]:
          vectormerge[k] = izq[i]
          i+= 1
        else:
          vectormerge[k] = der[j]
          j += 1
        k += 1

      # Nos fijamos si quedaron elementos en la lista
      # tanto derecha como izquierda

      while i < largo(izq):
        vectormerge[k] = izq[i]
        i += 1
        k += 1

      while j < largo(der):
        vectormerge[k] = der[j]
        j += 1
        k += 1
  merge(vectormerge)
  print("EL vector ordenado con merge es :", vectormerge)
mergesort(vectormerge)
```

Método de Ordenamiento Rápido (QuickSort)

Al igual que el ordenamiento por mezcla, el ordenamiento rapido es un algoritmo que lo llaman dive y venceras. El mismo funciona seleccionando un elemento como pivote y dividiendo la matriz dada alrededor del pivote elegido. Ha varios tipos de versiones de ordenamiento rápido que eligen los pivotes de maneras diferentes.

[Método de Ordenamiento Rápido (QuickSort)](https://www.youtube.com/watch?v=UrPJLhKF1jY)

```python
from random import sample
# Importamos un método de la biblioteca Random para generar una lista aleatoria

lista = list(range(100))# creamos la lista base con números del 1 al 100

# Creamos una lista aleatoria con sample
# (10 elementos aleatorios de la lista base)

vectorquick = sample(lista,10)
def quicksort(vectorquick, start = 0, end = len(vectorquick)-1):
  """Esta función ordena el vector que le pases como argumento con el metodo QuickSort"""

  #Imprimimos la lista obtenida al principio(Desordenada)
  print("El vector a ordenar con quick es:", vectorquick)

  def quick(vectorquick, start, end):
      if start >= end:
        return

      def particion(vectorquick, start, end):
          pivot = vectorquick[start]
          menor = start + 1
          mayor = end

          while True:
            # Si el valor actual es mayor que el pivot
            # esta en el lugar correcto(Lado derecho del pivote)y podemos
            # movernos hacia la izquierda, al siguiente elemento.
            # tambien debemos aegurarnos de no haber superado el puntero bajo, ya que indica
            # que ya hemos movido todos los elementos
            while menor <= mayor and vectorquick[mayor] >= pivot:
              mayor = mayor - 1 # Corrected: should decrement to move left

            # Proceso opuesto al anterior
            while menor <= mayor and vectorquick[menor] <= pivot:
              menor = menor + 1 # Corrected: should increment to move right


            # Encontramos un valor sea mayor o menor y que este fuera del arreglo
            #ó menor es mas grande que mayor, en cuya caso salimos del ciclo

            if menor <= mayor:
              vectorquick[menor],vectorquick[mayor]= vectorquick[mayor],vectorquick[menor]
              # Continua el bucle
            else:
              # Salimos del bucle
              break

          vectorquick[start], vectorquick[mayor]= vectorquick[mayor], vectorquick[start]

          return mayor # Corrected: particion must return the index of the pivot

      p = particion(vectorquick, start, end) # Corrected indentation
      quick(vectorquick, start, p-1)          # Corrected indentation
      quick(vectorquick, p+1, end)            # Corrected indentation

  quick(vectorquick, 0, len(vectorquick)-1) # Corrected indentation: initial call to the inner quick function
  print("El vector ordenado con el método quickSort es: ", vectorquick) # Corrected indentation

quicksort(vectorquick)
```

Método de Ordenamiento del Montón (HeapSort)

El método de ordenamiento del montón es similar a la clasificación por selección, donde primero encontramos el elemento máximo y lo colocamos al final. repetimos el mismo proceso para el resto de los elementos. Pero en el método debemos realizar previamente montone que son los que iran acomodandose con el algoritmo segun cual es mas grande de un lado o del otro del montón, y al mismo tiempo, se iran eliminando y acomodando los elementos mayores en la lista.

```python
#Código del Método de ordenamiento por montón

from random import sample

lista = list(range(100))

vectorheap = sample(lista,12)

def heapsort(vectorheap):
  """Esta función ordena el vector que le pases como argumento con el método HeapSort"""

  print("El vector a ordenar con el método heap es: ", vectorheap)

  largo = 0 # Establecemos un contador de largo

  for _ in vectorheap:
    largo += 1 #Obtenemos el largo del vector

  # Para amontonar la subparte a partir de i.
  # n es el tamaño del montón.

  def amontonar(vectorheap, n, i):
     mas_largo = i # tomamos i como el mas grande
     izq = 2*i+1
     der = 2*i+2

     if izq < n and vectorheap[i]< vectorheap[izq]:
          mas_largo = izq

    # Ver si existe la subparte de i correctamente y
    # Si es mayor que i

     if mas_largo != i:
          vectorheap[i],vectorheap[mas_largo] = vectorheap[mas_largo],vectorheap[i]
    # Vamos a cambiar el origen si es necesario
    # amontonar el origen
          amontonar(vectorheap, n, mas_largo)

  def heap(vectorheap):

    n = largo
    #Crear un montón maximo
    for i in range(n//2 - 1, -1, -1):
        amontonar(vectorheap, n, i)

    # Vamos a extraer elementos uno a uno
    for i in range(n-1, 0, -1):
        vectorheap[i], vectorheap[0] = vectorheap[0], vectorheap[i]
        # Intercambio
        amontonar(vectorheap, i, 0)

  heap(vectorheap)
  print("El vector ordeando con el método heap es:", vectorheap)

heapsort(vectorheap)
```

Método de ordenamiento del peine(ComboSort)
Este método es una mejora del método de ordenamiento burbuja, ya que en el método de la burbuja siempre se comparan valores consecutivos, entonces todos los cambios se realizan uno por uno. el método de peine mejora al bubble sort usando un espacio de tamaño superior a 1. El espacio comienza con un valor grande y se reduce a un factor x en cada iteración hasta que alcanza el valor 1. Por lo tanto este elimina más de una inversión con un intercambio y funciona mejor que el método de burbuja.

```python
#Codigo del algoritmo de Ordenamiento Comb Sort

from random import sample

lista = list(range(100))

vectorcomb = sample(lista,15)
def combsort(vectorcomb):
  """Esta función ordena el vector que le pases como argumento con el método de
  Comb Sort"""

  print("El vector a ordenar con el método comb es: ", vectorcomb)

  largo = 0 # Establecemos un contador del vector largo

  for _ in vectorcomb:
    largo += 1

  # Comencemos con la diferencia o distancia igual al largo del vector
  diferencia = largo

  # Establecemos la variable que define si es necesario o no
  # Intercambiar los números que se estan comparando

  cambiar = True

  while diferencia > 1 or cambiar:
    diferencia = max(1, int(diferencia / 1.25))
    # La diferencia minima es 1
    # en cada iteración vamos bajando la diferencia
    cambiar = False
    for i in range(largo - diferencia):
        j = i + diferencia
        # Ubicamos el número que esta a la distancia x de i
        if vectorcomb[i]<vectorcomb[j]:
            vectorcomb[i],vectorcomb[j]= vectorcomb[j], vectorcomb[i]
            # hacemos el intercambio de los números
            cambiar = True

  print("El vector ordenado con el método comb sort es: ", vectorcomb)

combsort(vectorcomb)
```

Método de Ordenamiento de la burbuja bidireccional (CocktailSort)
El método de ordenamiento CocktailSort es una reforma del método de ordenamiento de la buebuja, con la diferencia que en este el ordenamiento se realiza en ambas direcciones, comprobando que el primero sin ordenar sea el menor y el último sin ordenar sea el mayor y acomodando asi en su respectivo lugar cada uno de los elementos.
Método Burbuja Bidireccional

```python
#Algoritmo de Ordenamiento burbuja bidireccional

from random import sample

lista = list(range(100))

vectorcocktail = sample(lista,12)

def cocktailsort(vectorcocktail):
  """Esta función ordena el vector que le pases como
  argumento con el Método de Cocktail sort"""

  print("El vector a ordenar con el método cocktail es: ", vectorcocktail)

  largo = 0 # Establecemos un contador del largo

  for _ in vectorcocktail:
    largo += 1 # obtenemos el largo del vector

  for i in range(largo//2):# Comenzamos desde la mitad aproximadamente
      cambiar = False
      # Declaramos la variable que indica si es necesario intercambiar o no
      for j in range(1+i, largo-i):
        # Probar si los dos elementos estan en el orden incorrecto
          if vectorcocktail[j] < vectorcocktail[j-1]:
            # Entonces ambos elementos cambian de lugar
              vectorcocktail[j], vectorcocktail[j-1] = vectorcocktail[j-1], vectorcocktail[j]
              cambiar = True

            # si no ocurren cambios salimos del bucle
      if not cambiar:
        break
      cambiar = False
      for j in range(largo-i-1, i, -1):
        # Probar si los dos elementos estan en el orden incorrecto
          if vectorcocktail[j]<vectorcocktail[j-1]:
            # entonces ambos elementos cambian de lugar
              vectorcocktail[j],vectorcocktail[j-1]=vectorcocktail[j-1],vectorcocktail[j]
              cambiar = True
      if not cambiar:
        break
  print("El vector ordenado con el método cocktail es: ", vectorcocktail)

cocktailsort(vectorcocktail)
```

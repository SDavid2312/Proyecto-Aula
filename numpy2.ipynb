```python
#vamos a crear una matirz 3x3 con datos aleatorios y vamos a sacar el maximo, el minimo, .size, .ndim
#imprimir resultados

import numpy as np

m1 = np.array([2,3,4,5,6])
m2 = np.array([11,23,43,65])
print(np.concatenate((m1,m2)))
print(" ")

m3 = np.concatenate((m1,m2))
print(m3)
print(" ")
print("valor maximo de la matriz m3 :", m3.max())
print(" ")
print("valor minimo de la matriz m3 :", m3.min())

print("numeros de componentes de la matriz :", m3.size)
print(" ")
print("dimensionalidad de la matriz :", m3.ndim)
```

```python
#Solución de los sistemas matriciales

# AX = B - Soluciones gaussianas

A = np.array([[2,1,-2],[3,0,1],[1,1,-1]])# Matriz con los coeficientes de las ecuaciones
print(A)
print(" ")
B = np.array([[-3],[5],[2]]) # la matriz B
print(B)
print("")
print(np.transpose(B))
print(" ")
X = np.linalg.solve(A,B)
print(X)
```

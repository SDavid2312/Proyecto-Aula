```
import matplotlib.pyplot as plt #importamos la libreria para graficar
import numpy as np
%matplotlib inline
```

```
#vamos a empezar creando figuras lineales
fig, ax = plt.subplots()#crear una figura en el sitema de ejes X y Y
ax.plot([1,2,3,4],[1,4,2,3])#con el .plot voy a trabajar graficos de linea
plt.show()
```

```
x = np.linspace(0,10,100)#cree los datos y los asigne a una varibale
plt.plot(x, x, label = "linea")#indico que es un grafico lineal y le pongo una etiqueta
plt.legend()#con . legend muestro el texto en el grafico
plt.title("Grafica de linea recta")
plt.show()
```

```
#vamos a crear una grafica de linea pero con caracteristicas

flg = plt.figure()

ax = flg.add_subplot(111)#crea una figura en la fila 1, columna 1 y figura 1
ax.plot([2,4,6,8],[10,20,25,30], color = "lime", linewidth = 3)
ax.scatter([0.6,6.8,2.2,5.4],[11,25,9,26], color = "#FE420F", marker = "*")
plt.title("grafica combinada de linea con puntas de dispercion")
ax.set_xlim(0,10)
ax.set_ylim(0,40)
plt.show()
```

```
#vamos a hacer un grafico con superposiciones de diferentes graficos
plt.plot(x,np.sin(x) + np.cos(x), "r--", x, np.sin(x), "bs", x, np.cos(x), "g^")
plt.show()
```

```
x = np.linspace(0,2,100)
fig,ax = plt.subplots()#creamos una instancia de figura y una zona de figura
ax.plot(x, x, label = "funcion lineal")
ax.plot(x,x**2, label = "funcion cuadratica")
ax.plot(x, x**3, label = "Funcion cubica")
ax.set_xlabel("eje x")
ax.set_ylabel("eje y")
ax.set_title("Superposicion de tres funciones")
ax.legend()
plt.show()
```

Graficas de dispercion

```
#vamos a iniciar definiendo una base de datos aleatoria

data = {}
data = {"eje x": np.arange(50),
        "tamao": np.random.rand(50),
        "color": np.random.randint(0,50,50)}

data["eje y"] = data["eje x"] + 10*np.random.randn(50)
data["tamao"] = np.abs(data["tamao"])*100
```

```
data
```

```
data ["eje x"]
```

```
data ["eje y"]
```

```
#vamos a graficar la dispersion

plt.scatter("eje x","eje y", data = data)
plt.xlabel("valores guardados en a")
plt.ylabel("valores guardados en b")
plt.show()
```

```
#vamos a inducir la variacion de los tamaos en las burbujas de dispercion

plt.scatter("eje x", "eje y", s = "tamao", data = data)
plt.xlabel("valores guardados en a")
plt.ylabel("valores guardados en b")
plt.show()
```

```
#vamos a cambiar los colores de las esferas
plt.scatter("eje x", "eje y", s = "tamao", c = "color", cmap = "Dark2", data = data)
plt.xlabel("valores guardados en a")
plt.ylabel("valores guardados en b")
plt.show()
```

```
x = ["Maria","Pedro","juan"]
y = [15,18,25]
plt.bar(x,y)
plt.title("aos de cada persona")
plt.show()
```

```
x = ["Maria","Pedro","juan"]
y = [15,18,25]
plt.barh(x,y, color = "k")
plt.title("aos de cada persona")
plt.show()
```

```
# graficas de barras apilados

labels = ["G1","G2","G3","G4","G5"]
hombres = [24,26,30,35,17]
mujeres = [22,31,44,18,30]
width = 0.35

fig, ax = plt.subplots()

ax.bar(labels, hombres, width, label = "hombres")
ax.bar(labels, mujeres, width, bottom = hombres, label = mujeres)

ax.set_ylabel("Puntuaxiones")
ax.set_title("puntuaciones por grupo y genero")
ax.legend()
plt.show()
```

```
#miremos como creas un grafico de barras agrupado

x = np.arange(len(labels))
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, hombres, width, label = "hombres")
rects2 = ax.bar(x + width/2, mujeres, width, label = "mujeres")
ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.set_ylabel("puntuacion")
ax.set_title("puntuaciones por grupo y generos")
ax.legend()
plt.show()
```

```
from re import X
from matplotlib.lines import lineStyles
from matplotlib import patches
#como creamos un histograma
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(100000)

n, bins, patches = plt.hist(x, 30, density = True, facecolor = "b", alpha = 0.5, edgecolor = "black")
plt.axvline(x.mean(), color = "red", linestyle = "dashed", linewidth = 1.5, label = "mean")

plt.xlabel("puntaje")
plt.ylabel("probabilidad")
plt.title("histograma")
plt.axis([40, 160, 0, 0.03])
plt.legend()
plt.grid(False)#muestra cuadriculas de fondo
plt.show()
```

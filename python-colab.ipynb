```python
#vamos a ver otra forma de crear matrices mas rapido
m = np.arange(20).reshape(4,5)#arange(n componentes), el reshape es la dimensionalidad(# filas x # columnas)
print(m)
```

```python
m1 = np.arange(30).reshape(6,5)
print(m1)
print("dimencionalidad: ",m1.shape) #sacamos la dimensionalidad
print(" ")
print("diension: ",m1.ndim)
print(" ")
print("numero de componentes de la matriz: ",m1.size)
```

```python
#como podemos crear una mtriz nula o compuesta de ceros

m2 = np.zeros([3,4])
print(m2.size)
print(m2)
```

```python
# vamos a crear una matiz con datos aleatorios

m3 = np.linspace(0,100,25)#crear una matriz que va desde 0 a 100 y tiene 25 componentes
print(m3.size)
print(" ")
print(m3)
```

```python
#matrices de 3 dimensiones

m4 = np.arange(24).reshape(2,3,4)#creamos 2 matrices de dimencion 3x4
print(m4.size)
print(m4.ndim)
print(m4)
```

```python
m5 = np.arange(27).reshape(3,3,3)
print(m5.size)
print(m5.ndim)
print(m5)
```

```python
# hacer operaciones basicas con matrices

m6 = np.arange(24).reshape(4,6)
print(m6)
print(" ")
print(m6[3,4])#con esta operacion podemos extraer el dato de la fila n y la columna m

```

```python
#matiz de manera acendente
m7 = np.array([90,7,9,4,2,1])
print(m7.size)
print(np.sort(m7))#ordenamos la matriz de forma acendente
```

```python
#elevar una matiz a una potencia

print(np.power(m7,3))#con la funcion .power elevamos la matriz a un componente n
```

```python
#como aplicar un condicional dentro de una matriz

m8 = np.array([[23,54,67,78],[77,15,54,88]])
print(np.array(m8>=40))
print(m8.max())
print(m8.min())
```

```python
m1 = np.array([2,3,4,5,6])
m2 = np.array([11,23,43,65])
print(np.concatenate((m1,m2)))
print(" ")

m3 = np.concatenate((m1,m2))
print(m3)
print(" ")
print("valor maximo de la matriz m3 :", m3.max())
print(" ")
print("valor minimo de la matriz m3 :", m3.min())

print("numeros de componentes de la matriz :", m3.size)
print(" ")
print("dimensionalidad de la matriz :", m3.ndim)
```

```python
#Solucin de los sistemas matriciales

# AX = B - Soluciones gaussianas

A = np.array([[2,1,-2],[3,0,1],[1,1,-1]])# Matriz con los coeficientes de las ecuaciones
print(A)
print(" ")
B = np.array([[-3],[5],[2]]) # la matriz B
print(B)
print("")
print(np.transpose(B))
print(" ")
X = np.linalg.solve(A,B)
print(X)
```

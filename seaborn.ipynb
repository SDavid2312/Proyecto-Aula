displot

muestra la distribucion de un conjunto de univariante de observaciones

```python
#podemos realizar un histograma con los datos

sns.displot(tips["total_bill"]);
```

Se requiere eliminar la grafica y solo obtener el histograma

```python
sns.displot(tips["total_bill"], kde = True, bins = 30)
```

joinplot(): Permite emparejar los displot para datos bivariados. Con su selecciond con que tipo de parametro se va a comparar.

*   Escatter (dispercion=)
*   Reg
*   resid
*   Kde
*   hex



```python
# histograma y una dispercion

sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "scatter");
```

Vamos a realizar una combinacion de histograma con heaxgonos

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "hex");
```

Vamos a realizar un histograma con kde y scatter plot

```python
sns.jointplot(x = "total_bill", y = "tip", data= tips, kind = "reg");
```

pairplot : Grafica de relaciones por pares en un dataframe completo (para las columnas numericas) y soporta un argumento de tono de color(hue) para columnas categoricas

```python
#las diagonales son graficos de tipo histograma y los otros de tipo scatterplot

sns.pairplot(tips)# va a tomar los datos de tipo numericos
```

```python
#datos de tipos categoricos

#las graficas diagonales van a ser Kde y las otras graficas van a ser de tipo scatter plots

sns.pairplot(tips, hue = "sex", palette = "coolwarm");
```

```python
#vamos a crear el dataframe

df = sns.load_dataset("tips")
print(df.head())
```

miremos como se relacionan  los datos graficos de dispercion

```python
import matplotlib.pyplot as plt


sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "sex", style = "sex")
plt.title("relacion entre total bill y tip")
plt.show()
```

```python
sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "smoker", style = "smoker")
plt.title("relacion entre total bill y tip")
plt.show()
```

```python
sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "day", style = "day")
plt.title("relacion entre total bill y tip")
plt.show()
```

```python
sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "time", style = "time")
plt.title("relacion entre total bill y tip")
plt.show()
```

vamos a realizar el grafico de barras en seaborn

```python
sns.barplot(data = df, x = "day", y = "tip", hue = "sex", ci = None)
plt.title("Promedio de propinas de dia y genero")
plt.show()
```

```python
#grafico de cajas para realizar distribuciones

sns.boxenplot(data = df, x = "day", y = "total_bill", hue = "sex")
plt.title("Distribucion total por dia y genero")
plt.show()
```

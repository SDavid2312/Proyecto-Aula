Algoritmo de clasificacion - lirios

```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris #importamos los datos
```

Vamos a asignar la variable al dataset y luego lo visualizamos

```python
iris = load_iris()#Cargamos el data set a la variable
iris
```

```python
target_names = iris.target_names #es el comando que se usa para almacenar la informacion
features_names = iris.feature_names #almacenamiento de las caracteristicas
print(target_names)#nombre de los tipos de lirios
print(" ")
print(features_names)#Caracteristicas de los tipos de lirios
print(" ")
print(iris.target)
print(" ")
print(iris.data.shape)
```

```python
sepal_length = 0
fig, axs = plt.subplots(1,3, figsize = (15,4))

for ax, feature in zip(axs,[1,2,3]):
  for target, color in zip(range(3),["g","r","c"]):
    ax.scatter(iris.data[iris.target == target, feature], iris.data[iris.target == target, sepal_length],
                color = color, marker = "*", label = iris.target_names[target])
    ax.set_xlabel(iris.feature_names[feature], fontsize = 15)
    #la etiqueta del eje x y el tamaÃ±o de la letra
    ax.set_ylabel(iris.feature_names[sepal_length], fontsize = 15)
    ax.legend(prop = ({"size" : 10}))
    ax.grid(True)

plt.show()
```
